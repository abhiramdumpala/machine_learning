```python
def count_vowel_letters(input_string):
    vowel_letters = 'AEIOUaeiou'
    total_vowels = 0
    for character in input_string:
        if character in vowel_letters:
            total_vowels += 1
    return total_vowels

def count_consonant_letters(input_string):
    vowel_letters = 'AEIOUaeiou'
    total_consonants = 0
    for character in input_string:
        if character.isalpha() and character not in vowel_letters:  
            total_consonants += 1
    return total_consonants

input_string = input('Enter a String: ')
print('Count of vowels in the given string:', count_vowel_letters(input_string))
print('Count of consonants in the given string:', count_consonant_letters(input_string))
```

```python
def construct_matrix(num_rows, num_columns):
    matrix = []
    for row in range(num_rows):
        matrix_row = []
        for column in range(num_columns):
            element = int(input(f'Enter element for row {row + 1} and column {column + 1}: '))
            matrix_row.append(element)
        matrix.append(matrix_row)
    return matrix


def matrix_multiplication(matrix_a, matrix_b):
    rows_a = len(matrix_a)
    columns_a = len(matrix_a[0])
    rows_b = len(matrix_b)
    columns_b = len(matrix_b[0])
    
    result_matrix = [[0] * columns_b for _ in range(rows_a)]
    
    for row in range(rows_a):
        for column in range(columns_b):
            result_matrix[row][column] = sum(matrix_a[row][k] * matrix_b[k][column] for k in range(columns_a))
    
    return result_matrix


num_rows_a = int(input('Enter number of rows for matrix A: '))
num_columns_a = int(input('Enter number of columns for matrix A: '))

num_rows_b = int(input('Enter number of rows for matrix B: '))
num_columns_b = int(input('Enter number of columns for matrix B: '))

if num_columns_a != num_rows_b:
    print('Error: Multiplication is not possible')
    exit()

print('Enter elements for Matrix A:')
matrix_a = construct_matrix(num_rows_a, num_columns_a)

print('Enter elements for Matrix B:')
matrix_b = construct_matrix(num_rows_b, num_columns_b)

print('Input Matrix A:')
for row in range(num_rows_a):
    for column in range(num_columns_a):
        print(matrix_a[row][column], end=" ")
    print()

print('Input Matrix B:')
for row in range(num_rows_b):
    for column in range(num_columns_b):
        print(matrix_b[row][column], end=" ")
    print()

result_matrix = matrix_multiplication(matrix_a, matrix_b)

print('Result Matrix:')
for row in range(num_rows_a):
    for column in range(num_columns_b):
        print(result_matrix[row][column], end=" ")
    print()
```

```python
def input_elements():
    elements = []
    index = 0
    while True:
        user_input = input(f'Enter element {index + 1} (or type "exit" to finish): ')
        if user_input.lower() == "exit":
            break
        try:
            element = int(user_input)
            elements.append(element)
            index += 1
        except ValueError:
            print("Please enter a valid integer or 'exit' to finish.")
    return elements


def count_common_elements(list_one, list_two):
    common_count = 0
    for element_one in list_one:
        if element_one in list_two:
            common_count += 1
            list_two.remove(element_one)  
    return common_count


print('Enter elements for the first list:')
first_list = input_elements()

print('Enter elements for the second list:')
second_list = input_elements()

common_element_count = count_common_elements(first_list, second_list)
print(f'Number of common elements: {common_element_count}')
```

```python
def construct_matrix(num_rows, num_columns, matrix):
    for row in range(num_rows):
        matrix_row = []
        for column in range(num_columns):
            element = int(input(f'Enter element for row {row + 1} and column {column + 1}: '))
            matrix_row.append(element)
        matrix.append(matrix_row)


def transpose_matrix(original_matrix, num_rows, num_columns, transposed_matrix):
    for column in range(num_columns):
        transposed_row = []
        for row in range(num_rows):
            transposed_row.append(original_matrix[row][column])
        transposed_matrix.append(transposed_row)


num_rows = int(input('Enter number of rows: '))
num_columns = int(input('Enter number of columns: '))

matrix = []

construct_matrix(num_rows, num_columns, matrix)

print('Input Matrix:')
for row in range(num_rows):
    for column in range(num_columns):
        print(matrix[row][column], end=" ")
    print()

transposed_matrix = []

transpose_matrix(matrix, num_rows, num_columns, transposed_matrix)

print('Transpose Matrix:')
for row in range(num_columns):  
    for column in range(num_rows):  
        print(transposed_matrix[row][column], end=" ")
    print()
```
